```python
video_path =  "../data/video"
list_videos = [join(video_path,f) for f in os.listdir(video_path) if isfile(join(video_path,f)) 
               and f[-4:] == ".mp4"]
list_videos[0].split("/")[-1].split(".")[0]
```

```python
import numpy as np
from PIL import Image, ImageDraw
import cv2
from IPython.display import Image 
import torch
from torchvision import transforms
import torch.nn as nn

to_pil = transforms.ToPILImage()
to_tensor = transforms.ToTensor()
max_resizer = nn.AdaptiveMaxPool2d(output_size = (224,224))
        
for video_name in list_videos:
    print(video_name, "\n --- \n")
    cap = cv2.VideoCapture(video_name)

    videodims = (224,224)
    fourcc = cv2.VideoWriter_fourcc(*'avc1')    
    frame_rate = cap.get(cv2.CAP_PROP_FPS)
    video = cv2.VideoWriter("./all_224/{}_224.mp4".format(video_name.split("/")[-1].split(".")[0]),
                            fourcc, frame_rate, videodims)


    i = 0
    length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    while True:
        i += 1
        if i % 10 == 0:
            print(i, " / ", length, end = "\r")
        _, frame = cap.read()
        if frame is None:
            break
#         break
        frame = to_pil(cv2.cvtColor(np.array(frame), cv2.COLOR_RGB2BGR))
        old_size = frame.size # in (width, height) format
        central_crop = transforms.CenterCrop([min(old_size),min(old_size)])
        frame = central_crop(frame)
        new_frame = to_pil(max_resizer(to_tensor(frame)))
        video.write(cv2.cvtColor(np.array(new_frame), cv2.COLOR_RGB2BGR))
#         video.write(np.array(new_frame))
    print("done")
    video.release()
```
