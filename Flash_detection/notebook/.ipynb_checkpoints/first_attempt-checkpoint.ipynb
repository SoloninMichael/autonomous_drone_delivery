```python
args = {
    "video_dir": "../data/video/",
    "markup_dir": "../data/markup/",
    "std_height": 700,
    "std_width": 700,
    "in_channels": 3,
    "dropout": 0.3,
    "hidden_dim": 128,
}
args["video_path"] = Path(args["video_dir"])
args["markup_path"] = Path(args["markup_dir"])
```

```python
list_videos = [f for f in os.listdir(args["video_path"]) if isfile(join(args["video_path"],f))]
# names_prefix = [i.split(".")[0] for i in list_videos]

import random

list_videos
```

```python
to_standart = transforms.Compose([transforms.Resize((args["std_height"],args["std_width"]))])
to_pil = transforms.ToPILImage()
to_tensor = transforms.ToTensor()
```

```python
for idx, video_name in enumerate(list_videos):
    print("Reading {}-th video".format(idx))
    # get markup
    with open(args["markup_path"] / (video_name + ".markup.json")) as json_file:
        j = json.load(json_file)
        lst_tmp = j["target_pos"]
        target_points = torch.tensor([np.array(xi[1]) for xi in lst_tmp]) # frames_num, 2
    # get frames
    vid_path = str(args["video_path"] / (video_name))
    cap = cv2.VideoCapture(vid_path)
    frames_list = []
    while(cap.isOpened()):
        ret, frame = cap.read()
        if ret == False:
            break
        #TODO: make it optimal
        frame = to_pil(frame)
        frame = to_standart(frame)
        frame = to_tensor(frame)
        frames_list.append(frame)
    frames_tensor = torch.stack(frames_list)
```

```python
def get_batch(args, list_of_frames, list_of_targets):
    frames_tensor, f1,f2 = list_of_frames
    target_points, t1,t2 = list_of_targets
    # redo
    frames_lengths = [t.shape[0] for t in list_of_targets]
    max_len_batch = max(frames_lengths)
    for idx in range(len(list_of_frames)):
        to_append_shape = list(list_of_frames[idx].shape)
        to_append_shape[0] = max_len_batch - to_append_shape[0]
        to_append = torch.zeros(to_append_shape)
        list_of_frames[idx] = torch.cat([list_of_frames[idx], to_append], dim=0)
    for idx in range(len(list_of_targets)):
        to_append_shape = list(list_of_targets[idx].shape)
        to_append_shape[0] = max_len_batch - to_append_shape[0]
        to_append = torch.zeros(to_append_shape).long()
        list_of_targets[idx] = torch.cat([list_of_targets[idx], to_append], dim=0)
    return list_of_frames, list_of_targets
```

```python
l_frames, l_target = get_batch(args,[frames_tensor, f1,f2], [target_points, t1,t2])
```

```python
t_frames = torch.stack(l_frames)
```

```python
conv = nn.Sequential(
    nn.Conv2d(args["in_channels"], 4, kernel_size=(3,3), padding = (2,2), padding_mode="replicate"),
    nn.Conv2d(4, 8, kernel_size=(3,3)),
    nn.AvgPool2d(kernel_size=(3,3)),
    nn.Conv2d(8, 16, kernel_size=(3,3)),
    nn.MaxPool2d(kernel_size=(3,3)),
    nn.Conv2d(16, 16, kernel_size=(3,3)),
    nn.Conv2d(16, 32, kernel_size=(1,1)),
    nn.MaxPool2d(kernel_size=(3,3), stride=2),
    nn.Conv2d(32, 32, kernel_size=(3,3)),
    nn.Conv2d(32, 32, kernel_size=(3,3)),
)
fc1 = nn.Sequential(
    nn.Linear(2048, 512),
    nn.LeakyReLU(),
    nn.Dropout(p=args["dropout"]),
    nn.Linear(512, 256),
)
gru = nn.GRU(input_size=256, hidden_size=args["hidden_dim"], bidirectional=False, batch_first=True)
fc2 = nn.Sequential(
    nn.Linear(args["hidden_dim"], 32),
    nn.SELU(),
    nn.Linear(32,2)
)
```

```python
class  CharCNN(nn.Module):
    def __init__(self, args):
        super(CharCNN, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv1d(args["char_emb_dim"], 128, kernel_size=2, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=3, stride=2)
        )

        self.conv2 = nn.Sequential(
            nn.Conv1d(128, 64, kernel_size=2, stride=1),
            nn.ReLU(),
            nn.MaxPool1d(kernel_size=3, stride=1)
        ) 
        
        self.fc1 = nn.Sequential( 
            nn.Linear(320, 128),   ### fix param here
            nn.Sigmoid(),
            nn.Dropout(p=args["dropout"])
        )
        
        self.fc2 = nn.Sequential(
            nn.Linear(128, args["char_emb_out"]),
            #nn.ReLU(),
            #nn.Dropout(p=args["dropout"])
        )

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = x.view(x.size(0), -1)
        x = self.fc1(x)
        x = self.fc2(x)
        
        return x
```
